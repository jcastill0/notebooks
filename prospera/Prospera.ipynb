```python
import csv
import pandas as pd

MC_COMMUNITY = 'MC_Community.csv'
MC_GENERAL = 'MC_General.csv'
MC_SUBSCRIBERS = 'MC_Subscribers.csv'
SF_ACCOUNTS = 'SF_Account.csv'
SF_CONTACTS = 'SF_Contact.csv'
SF_LEADS = 'SF_Lead.csv'

mcCommDict = {}
with open('data/'+MC_COMMUNITY) as csvFile:
    csvReadH = csv.reader(csvFile, delimiter=',')
    for row in csvReadH:
        mcCommDict[str(row[0])] = MC_COMMUNITY
print (MC_COMMUNITY, len(mcCommDict))
        
mcGenDict = {}
with open('data/'+MC_GENERAL) as csvFile:
    csvReadH = csv.reader(csvFile, delimiter=',')
    for row in csvReadH:
        mcGenDict[str(row[0])] = MC_GENERAL
print (MC_GENERAL, len(mcGenDict))
        
mcSubDict = {}
with open('data/'+MC_SUBSCRIBERS) as csvFile:
    csvReadH = csv.reader(csvFile, delimiter=',')
    for row in csvReadH:
        mcSubDict[str(row[0])] = MC_SUBSCRIBERS
print (MC_SUBSCRIBERS, len(mcSubDict))
        
sfLeadDict = {}
lineCnt = 0
with open('data/'+SF_LEADS) as csvFile:
    csvReadH = csv.reader(csvFile, delimiter=',')
    for row in csvReadH:
        sfLeadDict[str(row[22])] = SF_LEADS
        lineCnt = lineCnt + 1
print (SF_LEADS, lineCnt)
print (SF_LEADS, len(sfLeadDict))
        
sfContactDict = {}
lineCnt = 0
with open('data/'+SF_CONTACTS) as csvFile:
    csvReadH = csv.reader(csvFile, delimiter=',')
    for row in csvReadH:
        sfContactDict[str(row[33])] = SF_CONTACTS
        lineCnt = lineCnt + 1
print (SF_CONTACTS, lineCnt)
print (SF_CONTACTS, len(sfContactDict))

sfAccountDict = {}
lineCnt = 0
with open('data/'+SF_ACCOUNTS) as csvFile:
    csvReadH = csv.reader(csvFile, delimiter=',')
    for row in csvReadH:
        sfAccountDict[str(row[33])] = SF_ACCOUNTS
        lineCnt = lineCnt + 1
print (SF_ACCOUNTS, lineCnt)
print (SF_ACCOUNTS, len(sfAccountDict))

commonLeadEmail = []
commonAcctEmail = []
commonContEmail = []
dictKeys = mcCommDict.keys()
with open('data/Common_MC_Community.csv', 'w+') as fH:
    for key in dictKeys:
        if (key in sfLeadDict):
            commonLeadEmail.append(key)
            fH.write(key + '\n')
        if (key in sfAccountDict):
            commonAcctEmail.append(key)
            fH.write(key + '\n')
        if (key in sfContactDict):
            commonContEmail.append(key)
            fH.write(key + '\n')
    print (MC_COMMUNITY, len(commonLeadEmail))
    print (MC_COMMUNITY, len(commonAcctEmail))
    print (MC_COMMUNITY, len(commonContEmail))

commonLeadEmail = []
commonAcctEmail = []
commonContEmail = []
dictKeys = mcGenDict.keys()
with open('data/Common_MC_General.csv', 'w+') as fH:
    for key in dictKeys:
        if (key in sfLeadDict):
            commonLeadEmail.append(key)
            fH.write(key + '\n')
        if (key in sfAccountDict):
            commonAcctEmail.append(key)
            fH.write(key + '\n')
        if (key in sfContactDict):
            commonContEmail.append(key)
            fH.write(key + '\n')
    print (MC_GENERAL, len(commonLeadEmail))
    print (MC_GENERAL, len(commonAcctEmail))
    print (MC_GENERAL, len(commonContEmail))

commonLeadEmail = []
commonAcctEmail = []
commonContEmail = []
dictKeys = mcSubDict.keys()
with open('data/Common_MC_Subscribers.csv', 'w+') as fH:
    for key in dictKeys:
        if (key in sfLeadDict):
            commonLeadEmail.append(key)
            fH.write(key + '\n')
        if (key in sfAccountDict):
            commonAcctEmail.append(key)
            fH.write(key + '\n')
        if (key in sfContactDict):
            commonContEmail.append(key)
            fH.write(key + '\n')
    print (MC_SUBSCRIBERS, len(commonLeadEmail))
    print (MC_SUBSCRIBERS, len(commonAcctEmail))
    print (MC_SUBSCRIBERS, len(commonContEmail))

print ('DONE')
```
