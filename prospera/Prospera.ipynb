```python
import csv
import pandas as pd

MC_COMMUNITY = 'MC_Community.csv'
MC_GENERAL = 'MC_General.csv'
MC_SUBSCRIBERS = 'MC_Subscribers.csv'
SF_CONTACTS = 'SF_Contact.csv'
SF_LEADS = 'SF_Lead.csv'

mcCommDict = {}
with open('data/src/'+MC_COMMUNITY) as csvFile:
    csvReadH = csv.reader(csvFile, delimiter=',')
    for row in csvReadH:
        mcCommDict[str(row[0])] = MC_COMMUNITY
print ('Number of records in '+ MC_COMMUNITY, len(mcCommDict))
        
mcGenDict = {}
with open('data/src/'+MC_GENERAL) as csvFile:
    csvReadH = csv.reader(csvFile, delimiter=',')
    for row in csvReadH:
        mcGenDict[str(row[0])] = MC_GENERAL
print ('Number of records in '+ MC_GENERAL, len(mcGenDict))
        
mcSubDict = {}
with open('data/src/'+MC_SUBSCRIBERS) as csvFile:
    csvReadH = csv.reader(csvFile, delimiter=',')
    for row in csvReadH:
        mcSubDict[str(row[0])] = MC_SUBSCRIBERS
print ('Number of records in '+ MC_SUBSCRIBERS, len(mcSubDict))
print ('**************')

sfLeadDict = {}
lineCnt = 0
with open('data/src/'+SF_LEADS) as csvFile:
    csvReadH = csv.reader(csvFile, delimiter=',')
    for row in csvReadH:
        sfLeadDict[str(row[22])] = row[0]
        lineCnt = lineCnt + 1
print ('Number of records in '+ SF_LEADS, lineCnt)
print ('Number of records in '+ SF_LEADS + ' with email', len(sfLeadDict))

sfContactDict = {}
lineCnt = 0
with open('data/src/'+SF_CONTACTS) as csvFile:
    csvReadH = csv.reader(csvFile, delimiter=',')
    for row in csvReadH:
        sfContactDict[str(row[33])] = row[0]
        lineCnt = lineCnt + 1
print ('Number of records in '+ SF_CONTACTS, lineCnt)
print ('Number of records in '+ SF_CONTACTS + ' with email', len(sfContactDict))
print ('******************')

commonLeadEmail = []
commonContEmail = []
dictKeys = mcCommDict.keys()
with open('data/results/Common_MC_Community.csv', 'w+') as fH:
    for key in dictKeys:
        if (key in sfLeadDict):
            commonLeadEmail.append(key)
            fH.write(key + '\n')
        if (key in sfContactDict):
            commonContEmail.append(key)
            fH.write(key + '\n')
    print ('Number of common records with Lead and '+ MC_COMMUNITY, len(commonLeadEmail))
    print ('Number of common records with Cntc and '+ MC_COMMUNITY, len(commonContEmail))

commonLeadEmail = []
commonContEmail = []
dictKeys = mcGenDict.keys()
with open('data/results/Common_MC_General.csv', 'w+') as fH:
    for key in dictKeys:
        if (key in sfLeadDict):
            commonLeadEmail.append(key)
            fH.write(key + '\n')
        if (key in sfContactDict):
            commonContEmail.append(key)
            fH.write(key + '\n')
    print ('Number of common records with Lead and '+ MC_GENERAL, len(commonLeadEmail))
    print ('Number of common records with Cntc and '+ MC_GENERAL, len(commonContEmail))

commonLeadEmail = []
commonContEmail = []
dictKeys = mcSubDict.keys()
with open('data/results/Common_MC_Subscribers.csv', 'w+') as fH:
    for key in dictKeys:
        if (key in sfLeadDict):
            commonLeadEmail.append(key)
            fH.write(key + '\n')
        if (key in sfContactDict):
            commonContEmail.append(key)
            fH.write(key + '\n')
    print ('Number of common records with Lead and '+ MC_SUBSCRIBERS, len(commonLeadEmail))
    print ('Number of common records with Cntc and '+ MC_SUBSCRIBERS, len(commonContEmail))
print ('******************')

newLeadUsers = []
dictKeys = sfLeadDict.keys()
with open('data/results/Not_in_MC_Leads.csv', 'w+') as fH:
    for key in dictKeys:
        if (key in mcCommDict):
            continue
        if (key in mcGenDict):
            continue
        if (key in mcSubDict):
            continue
        newLeadUsers.append(key)
        fH.write(key + ',' + sfLeadDict[key] + '\n')
    print ('Number of new Lead records not in MC', len(newLeadUsers))

newCntcUsers = []
dictKeys = sfContactDict.keys()
with open('data/results/Not_in_MC_Cntcs.csv', 'w+') as fH:
    for key in dictKeys:
        if (key in mcCommDict):
            continue
        if (key in mcGenDict):
            continue
        if (key in mcSubDict):
            continue
        newCntcUsers.append(key)
        fH.write(key + ',' + sfContactDict[key] + '\n')
    print ('Number of new Cntc records not in MC', len(newCntcUsers))

newMcUsers = []
with open('data/results/MC_Not_in_SF.csv', 'w+') as fH:
    dictKeys = mcCommDict.keys()
    for key in dictKeys:
        if (key in sfLeadDict):
            continue
        if (key in sfContactDict):
            continue
        newMcUsers.append(key)
        fH.write(key + '\n')
        
    dictKeys = mcGenDict.keys()
    for key in dictKeys:
        if (key in sfLeadDict):
            continue
        if (key in sfContactDict):
            continue
        newMcUsers.append(key)
        fH.write(key + '\n')

    dictKeys = mcSubDict.keys()
    for key in dictKeys:
        if (key in sfLeadDict):
            continue
        if (key in sfContactDict):
            continue
        newMcUsers.append(key)
        fH.write(key + '\n')

    print ('Number of new MC records not in SF', len(newMcUsers))

print ('DONE')
```
